```python
import gradio as gr
from transformers import pipeline
```

```python
sentiment = pipeline("sentiment-analysis")
```

```python
def get_sentiment(input):
  return sentiment(input)
```

```python
print("Enter Text = ")
txt = input()
print(get_sentiment(txt))
dic = get_sentiment(txt)[0]
print("Response = ",dic['label'])
```

```python
iface = gr.Interface(fn = get_sentiment,
                     inputs = "text" ,
                     outputs = ['text'],
                     title = "Sentiment Analysis",
                     description = "Get either positive or negative analysis from text")
```

```python
iface.launch()
```
